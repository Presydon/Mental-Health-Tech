# Mental Healh in the Tech

# **Background**

Data preparation plays a crucial role in analyzing and understanding attitudes towards mental health in the tech workplace. By collecting and organizing data, you can gain insights into how mental health is perceived within the tech/IT industry and the prevalence of mental health disorders among tech workers. The data preparation process for this survey involves several steps to ensure the data is clean, organized, and ready for analysis.

# **Overview**

The objectives of this project is to
  Predict Mental health attitudes, towards mental health in the tech workplace based on various factors in the dataset: This study should help identify the key factors that influence mental health attitudes and provide insights for organizations to develop targeted interventions and support system.

# **Data Exploration**

A brief overview of the data columns and what information they individually carry is as follows:

Timestamp: Time information was collected

Age: Demographic information

Gender: Demographic information

Country: Current geographical location

state: If you live in the United States, which state or territory do you live in?

self_employed: Are you self-employed?

family_history: Do you have a family history of mental illness?

treatment: Have you sought treatment for a mental health condition?

work_interfere: If you have a mental health condition, do you feel that it interferes with your work?

no_employees: How many employees does your company or organization have?

remote_work: Do you work remotely (outside of an office) at least 50% of the time?

tech_company: Is your employer primarily a tech company/organization?

benefits: Does your employer provide mental health benefits?

care_options: Do you know the options for mental health care your employer provides?

wellness_program: Has your employer ever discussed mental health as part of an employee wellness program?

seek_help: Does your employer provide resources to learn more about mental health issues and how to seek help?

anonymity: Is your anonymity protected if you choose to take advantage of mental health or substance abuse treatment resources?

leave: How easy is it for you to take medical leave for a mental health condition?

mental_health_consequence: Do you think that discussing a mental health issue with your employer would have negative consequences?

phys_health_consequence: Do you think that discussing a physical health issue with your employer would have negative consequences?

coworkers: Would you be willing to discuss a mental health issue with your coworkers?

supervisor: Would you be willing to discuss a mental health issue with your direct supervisor(s)?

mental_health_interview: Would you bring up a mental health issue with a potential employer in an interview?

phys_health_interview: Would you bring up a physical health issue with a potential employer in an interview?

mental_vs_physical: Do you feel that your employer takes mental health as seriously as physical health?

obs_consequence: Have you heard of or observed negative consequences for coworkers with mental health conditions in your workplace?

comments: Any additional notes or comments

## Data Cleaning

```python
# Drop columns that wouldn't be relevant for the analysis

drop_columns = ['Timestamp', 'state', 'comments']

df = df.drop(drop_columns, axis=1)
df.head()
```

```python
# Missing data indvidual columns and Total missing data. Before Refilling

individual_columns = df.isna().sum()
print(individual_columns)

group_missing = df.isnull().sum().sum()
print(f'\nThere are {group_missing} total missing data')
```

```python
# Handling the columns with missing data

print(f'{df.self_employed.value_counts()}')

print(f'\n{df.work_interfere.value_counts()}')
```

```python
"""
In order to fill the missing values, the 'Most Frequent Imputation' because the values are object and not integers.
This technique employs a method called mode imputation,
where missing values within a dataset are substituted with the most frequently occurring value (mode) from the available non-missing values within the corresponding column.
"""
# Import SimpleImputer from scikit-learn library
from sklearn.impute import SimpleImputer

df.work_interfere = np.ravel(SimpleImputer(strategy='most_frequent').fit_transform(df.work_interfere.values.reshape(-1, 1)))
df.self_employed = np.ravel(SimpleImputer(strategy='most_frequent').fit_transform(df.self_employed.values.reshape(-1, 1)))
print('Simple Imputer Done')
```

```python
# Missing data indvidual columns and Total missing data. After Refilling

individual_columns = df.isna().sum()
print(individual_columns)

group_missing = df.isnull().sum().sum()
print(f'\nThere are {group_missing} total missing data')
```

```python
# Removing duplicated data from the dataset

if df.duplicated().sum() != 0:
  print(f'There are {df.duplicated().sum()} duplicated values.')
  print('processing...')
  df.drop_duplicates(inplace=True)
  print(f'There are {df.duplicated().sum()} duplicate values')
```

```python
# Checking the columns to confirm their entries are valid
# Age Column
df.Age.unique()
```

```python
# Checking the columns to confirm their entries are valid
drop_rows = []

for index, row in df.iterrows():
  if row.Age < 18 or row.Age > 99:
    drop_rows.append(index)

df.drop(drop_rows, inplace=True)

df.reset_index(drop=True, inplace=True)

print(df.Age.unique())
print(f'\nThe youngest age is {df.Age.min()}, while the oldest age is {df.Age.max()}')
```

```python
# Checking the columns to confirm their entries are valid
# Gender Column
print(f'{df.Gender.unique()}')

print(f'\nThe number of Unique Genders are {df.Gender.nunique()}')
```

```python
""" For this project, we'd be grouping all 46 types into 3 (Male, Female and Others.) """

df.Gender.replace(['Male ', 'male', 'M', 'm', 'Male', 'Man', 'Mail',  'msle',
                   'Malr', 'Mal', 'maile', 'Make',], 'Male', inplace = True)

df.Gender.replace(['Female ', 'female', 'F', 'f', 'Woman', 'Female',
                     'femail', 'Femake', 'woman',], 'Female', inplace = True)

df.Gender.replace(['Female (trans)', 'queer/she/they', 'non-binary', 'Cis Male',
                     'fluid', 'queer', 'Androgyne', 'Trans-female', 'male leaning androgynous',
                      'Agender', 'A little about you', 'Nah', 'All', 'cis male',
                      'ostensibly male, unsure what that really means', 'Male-ish', 'Male (CIS)',
                      'Cis Man', 'Genderqueer', 'Enby', 'p', 'Neuter', 'something kinda male?',
                      'Guy (-ish) ^_^', 'Trans woman', 'Cis Female', 'cis-female/femme', 'Female (cis)',],
                  'Queer', inplace = True)

print(df.Gender.unique())
```
